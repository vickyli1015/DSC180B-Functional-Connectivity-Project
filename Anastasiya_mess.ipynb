```python
def euclidean(point1, point2):
    distance = 0
    for i in range(len(point1)):
        distance += (point1[i] - point2[i]) ** 2
    return (distance) ** 0.5
```

```python
rois = {
    "Left M1": [-38, -24, 62],
    "Right M1": [34, -22, 62],
    "Left SMA" : [-8, 0, 54],
    "Right SMA" : [8, 0, 54],
    "Left PMd": [-16, -20, 48],
    "Right PMd": [34, -8, 52]}
```

```python
all_points = set()
for k, v in rois.items():
    print(f"Closest point to {k}:")
    closest = find_closest_point(atlas_coords, v)
    all_points.add(closest)
```
